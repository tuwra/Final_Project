```python
import pandas as pd
import numpy as np
from numpy.linalg import norm

df = pd.read_csv("data/cleaned/cars_cleaned.csv")

price = df["price"].values

mean_price = np.sum(price) / len(price)

print("Mean price (manual):", mean_price)

std_price = np.sqrt(np.sum((price - mean_price) ** 2) / len(price))

print("Standard deviation:", std_price)

z_scores = (price - mean_price) / std_price

print("First 10 standardized values:")

print(z_scores[:10])

highest_price = df["price"].max()

lowest_price = df["price"].min()

print("Highest price:", highest_price)

print("Lowest price:", lowest_price)

highest = df[df["price"] == highest_price].iloc[0]

lowest = df[df["price"] == lowest_price].iloc[0]

a = highest[["odometer","year"]].values

b = lowest[["odometer","year"]].values

cos_sim = np.dot(a,b) / (norm(a)*norm(b))

print("Cosine similarity:", cos_sim)


# =========================
# Probability Estimate
# =========================

high_condition = df[df["condition"] == "excellent"]

probability = (high_condition["price"] > df["price"].mean()).mean()

print("Probability:", probability)
```
