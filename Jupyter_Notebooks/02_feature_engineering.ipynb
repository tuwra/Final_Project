```python
import pandas as pd
import numpy as np

df = pd.read_csv("Data/cleaned/cars_cleaned.csv")

current_year = 2024

df["car_age"] = current_year - df["year"]

df["price_per_km"] = df["price"] / df["odometer"]

manufacturer_count = df["manufacturer"].value_counts()

df["manufacturer_popularity"] = df["manufacturer"].map(manufacturer_count)

df.to_csv("data/cleaned/cars_cleaned.csv", index=False)

print("Feature engineering completed")
```
